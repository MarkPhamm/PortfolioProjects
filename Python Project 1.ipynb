```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

df = pd.read_csv(r'C:\Users\hamin\Downloads\movies.csv', index_col = 0)


```

```python
#let look at the data
df.head()
```

```python
#remove all missing data
df = df.dropna(how='any',axis=0)
```

```python
#Return Percentage of the
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
df.dtypes
```

```python
#Sorting gross value
df.sort_values(by = ['gross'], inplace = False, ascending = False).head()
```

```python
#Scatter plot with budget vs gross

plt.scatter(df.budget, df.gross)
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Films')
```

```python
# Plot the budget vs gross using seaborn

sns.regplot(x='budget', y = 'gross', data = df, scatter_kws = {"color": "red"}, line_kws = {"color":"blue"})
```

```python
#Plot the score and the gross using seaborn
sns.regplot(x='score', y = 'gross',data = df,scatter_kws = {"color": "red"}, line_kws = {"color":"blue"} )
```

```python
#Start looking at correlation using #pearson method
df.corr(method = 'pearson') #pearson, kendall, spearman
```

```python
# High Correlation Between Budget and Gross
```

```python
#Create correlation heatmap matrix
correlation_matrix = df.corr(method = 'pearson')

sns.heatmap(correlation_matrix, annot = True)
plt.title("Correlation matrix for Numeric Features")
plt.xlabel("Movie features")
plt.ylabel("Movie features")
plt.show()
```

```python
# Looking at the top 15 compaies by gross revenue

CompanyGrossSum = df.groupby('company')[["gross"]].sum()

CompanyGrossSumSorted = CompanyGrossSum.sort_values('gross', ascending = False)[:15]

CompanyGrossSumSorted = CompanyGrossSumSorted['gross'].astype('int64') 

CompanyGrossSumSorted
```

```python
#Look at company year and sum of gross profit
```

```python
#Look at company year and sum of gross profit
#Top 15 gross profit
CompanyGrossSum = df.groupby(['company', 'year'])[["gross"]].sum()

CompanyGrossSumSorted = CompanyGrossSum.sort_values(['gross','company','year'], ascending = False)[:15]

CompanyGrossSumSorted = CompanyGrossSumSorted['gross'].astype('int64') 

CompanyGrossSumSorted
```

```python
#Heat map between every correlation
correlation_matrix = df.apply(lambda x: x.factorize()[0]).corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Movies")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
#Correlation pair using unstack()
correlation_mat = df.apply(lambda x: x.factorize()[0]).corr()

corr_pairs = correlation_mat.unstack()
```

```python
#sorted the pairs
sorted_pairs = corr_pairs.sort_values(kind="quicksort")
```

```python
#print out the high correlation
high_correlation = sorted_pairs[abs(sorted_pairs) > 0.5]
print(high_correlation)
```
